## That dataframe

```python
# Splitting the datetime column into year, month, and hour to take advantage of the time data
# example: consider the # of bikers riding at 2AM vs. 2PM...

def parseDateTime(dtstr):
    """ Given a datetime string in the bikeshare dataset, returns the year, month, and hour of that data point. """
    year = dtstr[0:4]
    month = dtstr[5:7]
    hour = dtstr[-8:-6]
    return year, month, hour

def insertDateTimeFeatures(df):
    # getting values and putting them into years, months, hours for all the data points
    years, months, hours = zip(*(map(parseDateTime, df["datetime"].values)))

    # inserting things into the dataframe
    df.insert(1, 'year', pd.Series(years))
    df.insert(1, 'month', pd.Series(months))
    df.insert(1, 'hour', pd.Series(hours))
    
    return df

df = insertDateTimeFeatures(df)

# # since we now have year, month, and hour columns, we don't need the datetime column in the df (for now)
# df.drop('datetime', 1, inplace=True)

df
```
