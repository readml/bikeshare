```python
# Splitting the datetime column into year, month, and hour to take advantage of the time data
# example: consider the # of bikers riding at 2AM vs. 2PM...

def parseDateTime(dtstr):
    """ Given a datetime string in the bikeshare dataset, returns the year, month, and hour of that data point. """
    year = dtstr[0:4]
    month = dtstr[5:7]
    hour = dtstr[-8:-6]
    return year, month, hour

# getting values and putting them into years, months, hours for all the data points
years, months, hours = zip(*(map(parseDateTime, df["datetime"].values)))

# inserting things into the dataframe
df.insert(1, 'year', pd.Series(years))
df.insert(1, 'month', pd.Series(months))
df.insert(1, 'hour', pd.Series(hours))

# since we now have year, month, and hour columns, we don't need the datetime column in the df (for now)
df.drop('datetime', 1, inplace=True)

df
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# inputs
X = df.values[:,1:-3] # ignore datetimes, and casual, registered, and total count outputs
```

```python
# outputs
# assumes that casual and registered users follow different models (to get the total, simply sum casual and registered)
y_casual = df.values[:,-3]
y_registered = df.values[:,-2]
y_total = df.values[:,-1]
```

```python
# train test splitting
X_train, X_test, y_casual_train, y_casual_test, y_registered_train, y_registered_test, y_total_train, y_total_test = train_test_split(X, y_casual, y_registered, y_total, test_size = 0.3)
```

```python
# instantiate RF regressors for casual and registered bikers
rf_casual = RandomForestRegressor(n_jobs=-1)
rf_registered = RandomForestRegressor(n_jobs=-1)
```

```python
# fitting for the casual and registered bikers
%%time
rf_casual.fit(X_train, y_casual_train)
rf_registered.fit(X_train, y_registered_train)
```

```python
# predictions for the casual and registered bikers
y_casual_pred = rf_casual.predict(X_test)
y_registered_pred = rf_registered.predict(X_test)
```

```python
# total predictions is predicted # of casual bikers + predicted # of registered bikers 
y_total_pred = y_casual_pred + y_registered_pred
```

```python
# mean average error things
print "MAE for bikes rented per hour by casual riders: {}".format(np.mean(np.abs(y_casual_pred - y_casual_test)))
print "MAE for bikes rented per hour by registered riders: {}".format(np.mean(np.abs(y_registered_pred - y_registered_test)))
print "MAE for bikes rented per hour by all riders: {}".format(np.mean(np.abs(y_total_pred - y_total_test)))
```
