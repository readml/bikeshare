## That dataframe

```python
# Splitting the datetime column into year, month, and hour to take advantage of the time data
# example: consider the # of bikers riding at 2AM vs. 2PM...

def parseDateTime(dtstr):
    """ Given a datetime string in the bikeshare dataset, returns the year, month, and hour of that data point. """
    year = dtstr[0:4]
    month = dtstr[5:7]
    hour = dtstr[-8:-6]
    return year, month, hour

# getting values and putting them into years, months, hours for all the data points
years, months, hours = zip(*(map(parseDateTime, df["datetime"].values)))

# inserting things into the dataframe
df.insert(1, 'year', pd.Series(years))
df.insert(1, 'month', pd.Series(months))
df.insert(1, 'hour', pd.Series(hours))


# # since we now have year, month, and hour columns, we don't need the datetime column in the df (for now)
# df.drop('datetime', 1, inplace=True)

df
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.cross_validation import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
class bikesharePredictor:
    # TODO: resolve the battle between camelCase and underscores
    def __init__(self, df):
        self.df = df

        # TODO: Should we keep X1 and X2 in this class definition, or create an instance of bikesharePredictor and then
        # use the class methods as desired??
        # inputs to be used by predictor 
        self.X1 = self.firstPassX()
        self.X2 = self.secondPassX()

        # outputs to be used by predictor
        # assumes that casual and registered users follow different models (to get the total, simply sum casual and registered)
        self.y_casual = df.values[:,-3]
        self.y_registered = df.values[:,-2]
        self.y_total = df.values[:,-1]
        
        # TODO: Maybe have an array of these to iterate through for comparing results?
        print 'results for the first pass'
        self.y_pred1 = self.predict(self.X1)
        print '\nresults for the second pass'
        self.y_pred2 = self.predict(self.X2)


    def firstPassX(self):
        # First attempt at X for the predictive model using most columns of the df
        # inputs: self (should already have a dataframe)
        # output: x1, 1 for "first pass"
        return self.df.values[:,1:-3] # ignore datetimes, and casual, registered, and total count outputs

    def secondPassX(self):
        # Second pass takes advantage of separate date variables
        # inputs: self (should already have a dataframe)
        # output: x2, 2 for "second pass" --> should be in a format that can be used with scikit-learn's RF

        # Categorical data
        column_labels = {'hour':1, 'month':2, 'year':3, 'season':4, 'weather':7}
        cat_data = self.df.values[:, sorted(column_labels.values())]

        # Prepping categorical data for scikit-learn
        ohe = OneHotEncoder()
        ohe.fit(cat_data)
        cat_data_prepped = ohe.transform(cat_data)

        # Extract numerical data from df
        # Combine cat_data_prepped and numerical data to make a giant matrix to feed into the ML estimator
        numerical_data_labels = np.array([5, 6, 8, 9, 10, 11])
        numerical_data = df.values[:, numerical_data_labels]

        # Entire data matrix
        return np.column_stack((cat_data, numerical_data))

    def predict(self, X):
        # train test splitting
        X_train, X_test, y_casual_train, y_casual_test, y_registered_train, y_registered_test, y_total_train, y_total_test = train_test_split(X, 
                                                                                                                                              self.y_casual, 
                                                                                                                                              self.y_registered, 
                                                                                                                                              self.y_total, 
                                                                                                                                              test_size = 0.3)

        # TODO: Should these stay local or be class variables or??
        # instantiate RF regressors for casual and registered bikers
        rf_casual = RandomForestRegressor(n_jobs=-1)
        rf_registered = RandomForestRegressor(n_jobs=-1)

        # fitting for the casual and registered bikers
        rf_casual.fit(X_train, y_casual_train)
        rf_registered.fit(X_train, y_registered_train)

        # predictions for the casual and registered bikers
        y_casual_pred = rf_casual.predict(X_test)
        y_registered_pred = rf_registered.predict(X_test)

        # total predictions is predicted # of casual bikers + predicted # of registered bikers 
        y_total_pred = y_casual_pred + y_registered_pred
        
        # printing MAE for now
        # TODO: put all of the y variables into a dictionary, save as a class variable
        # TODO: make additional class methods for visualizing performances/model comparisons
        print "MAE for bikes rented per hour by casual riders: {}".format(np.mean(np.abs(y_casual_pred - y_casual_test)))
        print "MAE for bikes rented per hour by registered riders: {}".format(np.mean(np.abs(y_registered_pred - y_registered_test)))
        print "MAE for bikes rented per hour by all riders: {}".format(np.mean(np.abs(y_total_pred - y_total_test)))

#         return {"y_total_pred":y_total_pred, 
#                 "y_casual_pred":y_casual_pred,
#                 "y_registered_pred":y_registered_pred}
        return rf_casual, rf_registered
    
    def visualizeAndCompare(self):
        # TODO: Decide on some metrics for comparing performance and making modeling/design decisions
        pass
    
    def pickleResults(self):
        # TODO: Save results into a file for future use?
        pass
```
